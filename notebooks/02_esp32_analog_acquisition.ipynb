# ESP32 Analogue Acquisition Test

This notebook checks the MyoBridge acquisition pipeline using a potentiometer before connecting the MyoWare sensor.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../data/sample/esp32_analog_test.csv")

data.head()

data["time_seconds"] = (
    data["timestamp_us"] - data["timestamp_us"].iloc[0]
) / 1_000_000

plt.figure(figsize=(12, 4))
plt.plot(data["time_seconds"], data["adc_value"])

plt.title("ESP32 Potentiometer Acquisition Test")
plt.xlabel("Time (seconds)")
plt.ylabel("ADC value")
plt.grid(True)

plt.show()
```

```python
recording_duration = (
    data["timestamp_us"].iloc[-1]
    - data["timestamp_us"].iloc[0]
) / 1_000_000

estimated_sampling_rate = (
    len(data) - 1
) / recording_duration

print("Recorded samples:", len(data))
print("Recording duration:", recording_duration)
print("Estimated sampling rate:", estimated_sampling_rate)
```
